```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
```

# Producing train and test data

```python
def make_2d_pos_data(N, T, sig_t, sig_o):
    int_x = npr.rand()
    int_y = npr.rand()
    
    v = npr.rand(N, 2)
    t = np.arange(0, T)
    
    pos = np.einsum('ij,k->ijk', v, t)    
    pos += sig_o*npr.randn(N, 2, T)
    
    return pos
```

```python
train = make_2d_pos_data(10, 10, 1, 0.1)
test = make_2d_pos_data(10, 10, 1, 0.1)
```
